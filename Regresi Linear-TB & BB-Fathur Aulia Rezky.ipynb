```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## DATAFRAME

```python
data = {'Nama': ['Yudho', 'Aulia', 'Rizky', 'Fathur'],
        'Tinggi': [164, 169, 154, 168],
        'Berat': [66, 67, 55, 53]}

df = pd.DataFrame(data)
```

Program dimulai dengan membuat DataFrame menggunakan library pandas. DataFrame ini berisi data tinggi dan berat badan dari empat orang.

## Pembuatan Model

```python
#Model Data Uji
from sklearn.model_selection import train_test_split
x = df['Tinggi'].values.reshape(-1,1)
y = df['Berat'].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
#Model Data Regresi Linear
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

Selanjutnya, dilakukan pembuatan model regresi linear. Data dibagi menjadi data latih dan data uji menggunakan train_test_split, lalu model regresi linear didefinisikan dan dilatih menggunakan data latih.

## Prediksi pada Data Uji

```python
y_pred = model.predict(x_test)
print(f'Koefisien Regresi: {model.coef_[0]}')
```

Model yang telah dilatih digunakan untuk memprediksi berat badan berdasarkan data tinggi pada data uji. Koefisien regresi dari model juga dicetak.

## Visualisasi pada DataFrame

```python
plt.scatter(x, y, color='blue')
plt.plot(x, y, color='red', linewidth=3)
plt.title('Data Tinggi & Berat Badan')
plt.xlabel('Tinggi (cm)')
plt.ylabel('Berat Badan (kg)')
plt.show()
```

Dilakukan visualisasi data tinggi dan berat badan menggunakan scatter plot.

```python
tinggi_prediksi = [[168]]
berat_prediksi = model.predict(tinggi_prediksi)
print(f'Prediksi Berat Badan untuk Tinggi 168 cm : {berat_prediksi[0]}')
```

```python
berat_prediksi = [[53]]
tinggi_prediksi = model.predict(berat_prediksi)
print(f'Prediksi Tinggi Badan untuk Berat 53 kg : {tinggi_prediksi[0]}')
```

Terakhir, dilakukan prediksi individu. Dua contoh prediksi diberikan, yaitu prediksi berat badan untuk tinggi 168 cm, dan prediksi tinggi badan untuk berat 53 kg.
